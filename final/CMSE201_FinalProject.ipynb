## Methodology

```python
#reading excel sheet/data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

first_ed_stat = pd.read_excel (r'tabn330.10.xlsx', usecols = ("A:D"))
#renaming columns
ed_stat = first_ed_stat.rename(columns={first_ed_stat.columns[0]: 'Year', first_ed_stat.columns[1]: 'Total Cost', first_ed_stat.columns[2]: '4-year', first_ed_stat.columns[3]: '2-year'})
ed_stat
```

```python
#pulling lists
years = np.arange(1969,2019,1)
#years

all_cost = ed_stat['Total Cost']
#all_cost

#dividing by category
all_inst = all_cost[9:59]
#print(all_inst)
public = all_cost[65:115]
#print(public)
private = all_cost[121:171]
```

```python
#creating lists
all_fouryr = ed_stat['4-year']

#1970-79
four70_79 = all_fouryr[18] - all_fouryr[9]

#1980-89
four80_89 = all_fouryr[28] - all_fouryr[19]

#1990-99
four90_99 = all_fouryr[38] - all_fouryr[29]

#2000-09
four00_09 = all_fouryr[48] - all_fouryr[39]

#2010-19
four10_19 = all_fouryr[58] - all_fouryr[49]


#decades
decades = ["70-79", "80-89", "90-99", "00-09", "10-19"]
```

```python
#calculations
decades_4 = [four70_79, four80_89, four90_99, four00_09, four10_19]

abs_decades_4 = np.absolute(decades_4)
print(abs_decades_4)
```

## Results

```python
#public and private

plt.plot(years, public, 'lightgreen', label = "Public Institutions")

plt.subplot()
plt.plot(years, private, 'skyblue', label = "Private Institutions")

plt.xlabel("Year")
plt.ylabel("Cost (in current U.S. dollars)")
plt.title("Average College Costs in Recent Years")
plt.legend()
plt.grid()
```

```python
#all institutions
plt.plot(years, all_inst, 'salmon')
plt.xlabel("Years")
plt.ylabel("Cost (in current U.S. dollars)")
plt.title("Average College Costs for All Institutions in Recent Years")
plt.grid()
```

```python
#plotting
plt.plot(decades, abs_decades_4, 'o')
plt.grid()
plt.xlabel("Decade")
plt.ylabel("Price")
plt.title("Tuition Prices Changes Over Decades")
```

## Discussion and Conclusion

Overall, the price of all institutions have both increasing at a semi-consistent rate since the late 70s. There is a noticiable dip in the price of private institutions in the late 2010s, however these prices are still much higher than the public counterparts.
